```python
text=''
for line in open("Malibu.txt"):
    text += line.replace('\n','').replace('\xa3','').replace('\xba','')
#text = open("Malibu.txt").read().rstrip('\n')
```

```python
text
```

```python
from wordcloud import WordCloud
wordcloud = WordCloud().generate(text)
```

```python
%pylab inline
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
